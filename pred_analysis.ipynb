```python
lis = []
for idx,value in tqdm(data_total.iterrows()):
    for i in range(10):
        li = [value['singer'], value['song_name']]
        li.append(value[f'comment{i+1}'])
        lis.append(li)

comments_total = pd.DataFrame(lis,columns=['singer','song_name','caption'])
comments_total =  comments_total[~comments_total['caption'].isna()]
    
```

```python
def caption_length(string):
    if len(string) <2000:
        return string
    else:
        return ""
```

```python
data_merge = pd.merge(comments_total,data[['singer','song_name','tempo(category)','genre','emotion']], how='left',on=['singer','song_name'])
duple = data_merge[data_merge.duplicated('caption')]['caption'].values
data_merge = data_merge[~data_merge['caption'].isin(duple)]
data_merge = data_merge[~data_merge['caption'].isin(duple)]
data_merge['caption_length'] = data_merge['caption'].apply(len)
data_merge['caption'] = data_merge['caption'].apply(lambda x: caption_length(x))
data_merge = data_merge[data_merge['caption'] != ""]
# data_merge.to_csv('./comments_totals',index=False)
```
