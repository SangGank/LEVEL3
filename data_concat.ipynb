```python
import pandas as pd
import re
```

```python
data_1 =pd.read_csv('./genre_11_tempo_4.csv')
data_2 =pd.read_csv('./data_llama2.csv')
data_3 =pd.read_csv('./comments_process.csv')
data_4 = pd.read_csv('./when_data_process.csv')
data_3
```

```python
data_3 = data_3[data_3.caption_length <500]
data_3 = pd.merge(data_3,data_1[['singer', 'song_name','tempo(category)','genre','emotion']], how = 'left', on =['singer','song_name'])
data_3 = data_3[data_3['genre'].notna()]
data_3['caption'] = data_3['caption'].apply(lambda x : re.sub(r'[^A-Za-z0-9\'\!\?\.\s\t\n\,\"]', '', x))
duple = data_3[data_3.duplicated('caption')].caption.values
data_3 = data_3[~data_3.caption.isin(duple)]
data_3= data_3[data_3['genre'] != 'World']
data_3= data_3[data_3['genre'] != 'Pop']
data_3= data_3[data_3['genre'] != 'Rock']

```

```python
data_total = data_total[data_total['emotion'].notna()]
data_total['caption_length'] = data_total['caption'].str.len()
data_total = data_total[data_total['caption_length']>10]
```
