```python
import pandas as pd
```

```python
data_1 =pd.read_csv('./genre_11_tempo_4.csv')
data_2 =pd.read_csv('./data_llama2.csv')
data_3 =pd.read_csv('./comments_process.csv')
data_3
```

```python
data_3 = data_3[data_3.caption_length <500]
data_3 = pd.merge(data_3,data_1[['singer', 'song_name','tempo(category)','genre','emotion']], how = 'left', on =['singer','song_name'])
data_3[data_3.singer == "'s hands... the man was born to play keyboards."]
```

```python
duple = data_1[data_1.duplicated('caption')].caption.values
data_1 = data_1[~data_1.caption.isin(duple)]
data_1
```

```python
data_2 = data_2[data_2.genre != 'World']
```

```python
data_total = pd.concat([data_1[['caption','tempo(category)','genre','emotion']],data_2[['caption','tempo(category)','genre','emotion']], data_3[['caption','tempo(category)','genre','emotion']]])
data_total
```

```python
data_total.genre.value_counts()
```

```python
data_total = data_total[data_total['emotion'].notna()]
data_total.to_csv('./total_data.csv',index =False)
```
