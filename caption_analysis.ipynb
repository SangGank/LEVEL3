```python
data1 = pd.read_csv('./data_comments.csv')
data2 = pd.read_csv('./data_comments2_copy.csv')
data3 = pd.read_csv('./data_comments3.csv')
data_total = pd.concat([data1,data2,data3],ignore_index=True)
data_total = data_total[data_total['comment1'].notna()]
data_total
```

```python
lis = []
columns = ['singer','song_name','caption']
for idx, value in data_total.iterrows():
    for i in range(10):
        lis.append([value['singer'],value['song_name'],value[f'comment{i+1}']])
caption_data = pd.DataFrame(lis,columns=columns)
caption_data
```

```python
duple = caption_data[caption_data.duplicated('caption')]['caption'].values
caption_data = caption_data[~caption_data['caption'].isin(duple)]
caption_data
```

```python
caption_data['caption_length'] = caption_data.caption.str.len()
caption_data['language'] = caption_data['caption'].apply(lambda x: langid.classify(x)[0])
print(caption_data['caption'])
caption_data[(20<caption_data['caption_length']) & (caption_data['caption_length']<1000)]
```
